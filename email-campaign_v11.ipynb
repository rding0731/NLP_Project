```python
from sompy.sompy import SOMFactory
```

```python
from sklearn.preprocessing import LabelEncoder
# non_number_columns = train.dtypes[train.dtypes == object].index.values
non_number_columns=['Customer_Location']
for column in non_number_columns:
    train_test = pd.concat([train_data[column], test_data[column]], axis = 0)
    encoder = LabelEncoder().fit(train_test.astype(str))
    train_data[column] = encoder.transform(train_data[column].astype(str)).astype(np.int32)    
    test_data[column] = encoder.transform(test_data[column].astype(str)).astype(np.int32)
```

```python
test_df=train_data.fillna(0)
test_df=test_df.drop(['Email_ID'],axis=1)
test_df=test_df.sort_values(by='Email_Status',ascending=False)
test_df=test_df[['Email_Status','Email_Type', u'Subject_Hotness_Score', u'Email_Source_Type',
       u'Customer_Location', u'Email_Campaign_Type',
       u'Total_Past_Communications', u'Time_Email_sent_Category',
       u'Word_Count', u'Total_Links', u'Total_Images']]
names=test_df.columns.tolist()
test_df=test_df.as_matrix()
```

```python
msz0, msz1=120,120  # total number of neurals=msz0*msz1
ini_method='random'
# ini_method='pca'
sm = SOMFactory().build(test_df, mapsize = [msz0, msz1],normalization = 'var', initialization=ini_method, component_names=names)
sm.train(n_job=1, verbose=False, train_rough_len=2, train_finetune_len=5)
```

```python
topographic_error = sm.calculate_topographic_error()
quantization_error = np.mean(sm._bmu[1])
print ("Topographic error = %s; Quantization error = %s" % (topographic_error, quantization_error))
# Topographic error = 0.507146723626; Quantization error = 1.44601318638
```

```python
from sompy.visualization.mapview import View2D
view2D  = View2D(msz0,msz1,"rand data",text_size=10)
view2D.show(sm, col_sz=2, which_dim="all", desnormalize=True)
plt.show()
```
