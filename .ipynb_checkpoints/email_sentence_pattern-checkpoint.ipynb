## Group the subject by FORMAT and KEY-WORDS of the sentence in email subject

This script explores the email patterns:
    format 1: with ':' in the sentence, key words=the words before ":"
            Pattern: Hot words : Sentence
    format 2: with 'âˆ’' in the sentence, key words=the words before "âˆ’
            Pattern=Product - Sentence
    format 3: with '+' in the sentence,
            key words=the words before "+
            content=the words after +
            such as : Up to 75% OFF clearance + MORE SAVINGS when you open a credit card!
            +: means 'AND', Nothing special   
    format 4: with '|' in the sentence, key words=the words before "|"
            Nothing special, | means mutiple products, see below
    format 42: it seems we should try pattern with ...|...|...|, 
            They are a list of items/products, no relationship between items: Product1 | Product2 | Product3 |...
    format 5: with numbers
            Date, 50%, Dollar, ...
    CAPTIAL Words
         product type: ['NFL'],['FTC', 'ISO,'],['MLBâ„¢'],['FTC', 'ISO,'],
         Others:
             ['DOUBLE', 'REWARDS'],
             ['BREAKING:'],
             ['NEW'],
             ['YOU', 'BERNINA'],
             ['FIRST', 'ACCESS'],
             ['NEW!',
Last: find pattern, {KeyWord}{Date}{70%}{number}{product}
    Conclusion:
              We can apply pattern/format to pre-processing the data, and manually check, group them one-by-one

##### group by format:

```python
'''
brek senece by special sysmbol like :,-,|
keywords =words before special words
'''
def key_word_comma_extract(special_symbol,string):
    pattern=re.compile('.*?('+special_symbol+').*?')
    results=re.findall(pattern,string)
    symbol=special_symbol
    if('\\' in symbol):
        symbol=special_symbol[1:]
    ind=string.index(symbol)
    key_words=''
    content=''
    N=len(special_symbol)
    if(ind>0):
        key_words=string[0:ind]
        content=string[ind+N:]
    return key_words,content
def sentence_break_by_symbol(email_df,special_symbol):
    symbol=special_symbol
    if('\\' in symbol):
        symbol=special_symbol[1:]
    tmp=[True if symbol in x else False for x in email_df.SUBJECT]
    tmp_df=email_df[tmp]
    num=0
    for x in tmp_df.SUBJECT:
        key_words,content=key_word_comma_extract(special_symbol,x)
        df=pd.DataFrame({'Sentence':[x],'KeyWords':[key_words],'Content':[content]})
        if(num==0):
            grouped_df=df.copy()
        else:
            grouped_df=grouped_df.append(df)
        num+=1
    grouped_df=grouped_df.reset_index(drop=True)
    grouped_df=grouped_df[['KeyWords','Content','Sentence']]
    return grouped_df
```

### format 5: with numbers

```python
def key_word_digit_extract(string):
    
    key_words=[]
    pattern=re.compile('.*?([0-9]+).*?')
    results=re.findall(pattern,string)
    if(len(results)==0):
        return key_words
    words=string.split(' ')
    for word in words:
        result=re.findall(pattern,word)
        if(len(result)>0):
            key_words.append(word)
    return key_words


tmp=[False]*len(email_df)
num=0
for x in email_df.SUBJECT:
    key_words=key_word_digit_extract(x)
    if(len(key_words)>0):
        df=pd.DataFrame({'Sentence':[x],'KeyWrods':[key_words]})
        if(num==0):
            F5_df=pd.DataFrame({'Sentence':[x],'KeyWrods':[key_words]})
        else:
            F5_df=F5_df.append(df)
        num+=1
F5_df=F5_df.reset_index(drop=True)
# for i in range(len(F5_df)):
for i in range(16):    
    print (F5_df.loc[i,'KeyWrods'])
    print (F5_df.loc[i,'Sentence'])
```

# Pattern search using NLTK

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords, words
from langdetect import detect
import string
import sys

input_file = sys.argv[1]
output_file = sys.argv[2]

def is_english(text):
    if not isinstance(text, basestring):
        return False
    text = text.strip()
    text = ''.join([c for c in text if c not in string.punctuation])
    if len(text.split()) < 5:
        return True if all(t.lower() in words.words() for t in text.split()) else False
    return 'en' == detect(text)


CFG = r"""
    NBLK:
        {<NN.*><IN><NN.*>}  #Nouns followed by in/of/etc... terminated with Nouns
        {<NN.*|JJ>*<NN.*>}  # Nouns and Adjectives, terminated with Nouns
        
    NP:
        {<NBLK>}
        {<NBLK><IN><NBLK>}  # Above, followed by in/of/etc... terminated with NBLK
"""

stopwords = stopwords.words('english')

def leaves(tree):
    return [subtree.leaves() for subtree in tree.subtrees(filter = lambda t: t.label()=='NP')]

def normalize(word, tag):
    return word.lower() if tag != 'NNP' else word

def acceptable_word(word):
    return 2 <= len(word) <= 40 and word.lower() not in stopwords

def get_phrases(tree):
    return [' '.join([normalize(w,t) for w,t in leaf]) for leaf in leaves(tree)]

def get_phrases2(tree):
    name = []
    topic = []
    for leaf in leaves(tree):
        tok, tag =  zip(*[(w,t) for w,t in leaf])
        #print tok, tag
        #tok = ' '.join(tok)
        if all([tg=='NNP' for tg in tag]) and not any([tk.lower() in words.words() for tk in tok]):
            name.append(' '.join(tok))
        else:
            topic.append(' '.join(tok))
    return ','.join(name), ','.join(topic)

def get_phrase_wrapper(text, chunker):
    toks = nltk.word_tokenize(text)
    postoks = nltk.tag.pos_tag(toks)
    tree = chunker.parse(postoks)
    print(tree)
    return get_phrases2(tree)

chunker = nltk.RegexpParser(CFG)
# df = pd.read_csv(input_file, encoding='mac_roman')
# df['is_eng'] = df['cleaned_text'].apply(is_english)

# df = df[df.is_eng]
# df['names'],df['topics'] = zip(*df['cleaned_text'].apply(lambda x: get_phrase_wrapper(x, chunker)))
```

```python
def get_noun_phrases(text,chunker):
    toks = nltk.word_tokenize(text)
    postoks = nltk.tag.pos_tag(toks)
    tree = chunker.parse(postoks)    
    name = []
    topic = []
    for leaf in leaves(tree):
        tok, tag =  zip(*[(w,t) for w,t in leaf])
        name.append(' '.join(tok))
    return name

def string_normalize(s):
    if not s: return ''
    else:
#         for p in "!\".$%'()*,-/:;<=>?[\]^_`{|}~":
#             s = s.replace(p, ' ')
        s=s.replace("'",' ')
    return " ".join(s.split()).strip()
def date_find(string):
#     string = 'I want to apply for leaves from 12/12/2017 to 12/18/2017 I want to apply for leaves from 12 January 2017 to ' \
#        '12/18/2017 I want to apply for leaves from 12/12/2017 to 12 Jan 17, May 20, 2017'
    # 12/12/2017, 12 January 2017, and 12 Jan 17 are the only possible patterns 
    date_string=[]
    matches = re.findall('(\d{2}[\/ ](\d{2}|January|Jan|February|Feb|March|Mar|April|Apr|May|May|June|Jun|July|Jul|August|Aug|September|Sep|October|Oct|November|Nov|December|Dec)[\/ ]\d{2,4})', string)
    for match in matches:
        date_string.append(match[0])
    # May 20,2017 are the only possible patterns 
    matches = re.findall('((January|Jan|February|Feb|March|Mar|April|Apr|May|May|June|Jun|July|Jul|August|Aug|September|Sep|October|Oct|November|Nov|December|Dec|)[ ](\d{2})(, )\d{2,4})', string)
    for match in matches:
        date_string.append(match[0])
#     06.05.17
    matches = re.findall('(\d{2}(\.)\d{2}(\.)\d{2,4})', string)
    for match in matches:
        date_string.append(match[0])
    return date_string
```

```python
Noun_grammar = r"""
  NP: {<NNP>?<DT|PP\$>?<JJ>*<NN>+<NNS>?}   # chunk determiner/possessive, adjectives and noun
      {<NNP>+}                # chunk sequences of proper nouns
"""
chunker_noun = nltk.RegexpParser(Noun_grammar)
```

```python
i=13
i=15
i=14
i=16
i=17
i=18
i=19
KEY_words='credit card'
debug=False
for i in range(30):
    text0=F5_df.loc[i,'Sentence']
    text=string_normalize(text0)
    print('**************raw text ****************')
    print (i,text0)
    # print('---------- extracted pattern -----------------')
    # Step 1: replace KEY_words related Noun words to {KEY_words}
    noun_words=get_noun_phrases(text,chunker_noun)
    for word in noun_words:
        if(KEY_words in word.lower()):
            Noun_words=word
            if(debug):print ('KeyWords={}'.format(Noun_words))
#     text_new=re.sub(Noun_words,'{KEY_words}',text)
    text_new=text.replace(Noun_words,'{KEY_words}')
    if(debug): print('----------Step1: replace Key words -----------------')
    if(debug):
        print(text_new)
#     # Step 2: replace XX% with {%}
    words=text_new.split(' ')
    for (i,word) in enumerate(words):
        if('%' in word):
            words[i]='{%}'
    text_new=' '.join(words)
    if(debug): print('----------Step2: replace percentage words -----------------')
    if(debug):
        print(text_new)
    # Step 3: repalce Date with {Date}
    date_string=date_find(text_new)
    if(debug): print('----------Step3:  replace Date -----------------')
    if(debug):
        print(text_new)
    words=text_new.split(' ')
    for word in date_string:
        text_new=text_new.replace(word,'{Date}')
    if(debug):
        print(text_new)
    # Step 4: replace number to {number}
    digital_words=key_word_digit_extract(text_new)
    if(debug): print('----------Step4:  replace digital words -----------------')
    if(debug): print(digital_words)
    words=text_new.split(' ')
    for (i,word) in enumerate(words):
        if(word in digital_words):
            if('$' in word):
                words[i]='{$dollar}'
            else:
                words[i]='{number}'
    text_new=' '.join(words)
    print ('----pattern sentence------')
    print(text_new)
```

# Capital Letters and WORDS

```python
def capital_words_extraction(string):
    words=string.split(' ')
    capital_words=[]
    pattern=re.compile('([A-Z]+)()')
    for word in words:
        result=re.findall(pattern,word)
        if(len(result)>0):
            capital_words.append(word)
    return capital_words
def upper_words_extraction(string):
    words=string.split(' ')
    upper_words=[]
    pattern=re.compile('[A-Za-z]+')
    for word in words:
        if(word.upper()==word):
            result=re.findall(pattern,word)
            if((len(result)>0) & (len(word)>1)):
                upper_words.append(word)
    return upper_words
email_df['CapitalWords']=email_df.SUBJECT.apply(capital_words_extraction)
email_df['UpperWords']=email_df.SUBJECT.apply(upper_words_extraction)
UpperWords=[]
for x in email_df.UpperWords:
    if(len(x)>0):
        UpperWords.append(x)
UpperWords
```
